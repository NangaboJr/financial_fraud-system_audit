```python
import pandas as pd
df = pd.read_csv('Financial Analysis.csv')
print(df.head())
```

```python
fraud_by_type = df.groupby('type')['isFraud'].sum()
print("Fraud counts by Transaction Type:")
print(fraud_by_type)
print("\nSystem Flags vs Actual Fraud:")
print(pd.crosstab(df['isFraud'], df['isFlaggedFraud']))
```

```python
summary = pd.DataFrame({
    'Total': df['type'].value_counts(),
    'Fraud': df[df['isFraud'] == 1]['type'].value_counts()
}).fillna(0) # Replace NaN with 0 for types with no fraud
summary['Fraud_Rate_%'] = (summary['Fraud'] / summary['Total']) * 100

print(summary)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
df[df['isFraud'] == 0]['step'].hist(bins=100, label='Normal', alpha=0.5, color='blue')
df[df['isFraud'] == 1]['step'].hist(bins=100, label='Fraud', alpha=1, color='green')

plt.title('Transaction Volume over Time (Normal vs Fraud)')
plt.xlabel('Time Step (Hours)')
plt.ylabel('Number of Transactions')
plt.legend()
plt.yscale('log')
plt.show()
```

```python
emptied = df[df['amount'] == df['oldbalanceOrg']]

print(f"Total transactions that emptied the account: {len(emptied)}")
print(f"Percentage of those that were actually FRAUD: {(emptied['isFraud'].mean() * 100):.2f}%")
```

```python
missing_at_dest = df[(df['isFraud'] == 1) & (df['newbalanceDest'] == 0) & (df['amount'] > 0)]

print(f"Number of Fraud cases where money 'vanished' at destination: {len(missing_at_dest)}")
print(f"Percentage of total fraud where this happened: {(len(missing_at_dest) / df['isFraud'].sum() * 100):.2f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
risk_df = df.groupby('type')['isFraud'].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(10,6))
sns.barplot(x=risk_df.index, y=risk_df.values, palette='Reds_r')
plt.title('Percentage of Transactions that are FRAUD by Type', fontsize=15)
plt.ylabel('Fraud Probability (%)')
plt.xlabel('Transaction Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=risk_df.index, y=risk_df.values, hue=risk_df.index, palette='Reds_r', legend=False)
plt.title('Percentage of Transactions that are FRAUD by Type', fontsize=15)
plt.ylabel('Fraud Probability (%)')
plt.xlabel('Transaction Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
